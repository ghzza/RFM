# Import Data and Data Preprocessing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('Customer.csv')
transaction = pd.read_csv('Transactions.csv')
product = pd.read_csv('prod_cat_info.csv')
```

```python
#Renaming column "prod_sub_cat_code" to "prod_subcat_code" so that datasets merged easily Transaction and Product_heirarchy
product.rename(columns = {"prod_sub_cat_code":"prod_subcat_code"},inplace = True)
```

```python
product
```

```python
#Merging Datasets transaction and product_heirarchy with the help of left join and assigning it to variable Data
Data= pd.merge(left = transaction,
        right = product,
        on = ["prod_cat_code","prod_subcat_code"],
        how = "left",
        )
Data
```

```python
dfFinal = pd.merge(left = Data,
                         right = customers,
                         left_on = "cust_id",
                         right_on = "customer_Id",
                         how = "left")
dfFinal
```

```python
dfFinal.info()
```

```python
dfFinal.describe()
```

```python
dfFinal.describe(include='object')
```

```python
dfFinal.isnull().sum()
```

```python
dfFinal = dfFinal[dfFinal['Qty']>0]
dfFinal = dfFinal.reset_index(drop=True)
dfFinal
```

## Rename Rate column to Price 

```python
dfFinal.rename(columns={'Rate':'Price'}, inplace = True)
```

## Convert tran_date to datetime

```python
dfFinal['tran_date'] = pd.to_datetime(dfFinal['tran_date'])
dfFinal['tran_date'] = dfFinal['tran_date'].dt.date
```

```python
dfFinal.info()
```

```python
dfFinal
```

# Frequency and Monetary

```python
#max_date
max_date = dfFinal['tran_date'].max()
Cust_aggregate = dfFinal.groupby('cust_id').agg({'tran_date':[('Max Date','max'),('Frequency','count')],'Price':[("Monetary",'sum')]})
```

```python
Cust_aggregate.columns = Cust_aggregate.columns.droplevel(0)
Cust_aggregate = Cust_aggregate.reset_index()
```

```python
Cust_aggregate
```

# Recency

```python
Cust_aggregate['Recency'] = Cust_aggregate['Max Date'].apply(lambda x: (max_date - x).days)
Cust_aggregate['Recency']
```

```python
Cust_aggregate
```

# Age and Return

```python
Cust_Age = dfFinal.groupby('cust_id')['tran_date'].min().reset_index()
Cust_Age['Age'] = Cust_Age['tran_date'].apply(lambda x: (max_date-x).days/365)
Cust_Age
```

```python
Cust_Age.info()
```

```python
Cust_Age.describe()
```

```python
Cust_Return = dfFinal[['cust_id','tran_date']].drop_duplicates()
Cust_Return = Cust_Return[['cust_id','tran_date']].drop_duplicates().set_index('cust_id').diff()
Cust_Return['Nrow'] = Cust_Return.groupby('cust_id').cumcount()+1
```

```python
Cust_Return.info()
```

```python
Cust_Return
```

```python
Cust_Return.describe()
```

```python
Cust_Age_Return = pd.merge(Cust_Age, Cust_Return, on='cust_id')
Cust_Age_Return.columns = ['cust_id','tran_date','Age','Return','Nrow']
Cust_Age_Return
```

```python
Cust_Age_Return.describe()
```

```python
#merge with data cust_aggregate
Cust_aggregate = pd.merge(Cust_aggregate, Cust_Age_Return, on='cust_id', how='left')
```

```python
Cust_aggregate.isna().sum()
```

```python
Cust_aggregate
```

```python
Cust_aggregate.describe()
```

```python
#remove first transaction except for the only one transaction
Cust_aggregate = Cust_aggregate[((Cust_aggregate['Nrow']!=1) &(Cust_aggregate['Frequency']>1))] 
```

```python
Cust_aggregate.describe()
```

```python
Cust_aggregate['Return'] = Cust_aggregate['Return'].apply(lambda x: x.days)

Cust_aggregate.groupby('cust_id').agg({'Age':'min','Return':'median'}).reset_index()
```

```python
Cust_aggregate.describe()
```

```python
Cust_aggregate[['cust_id','tran_date','Return']].groupby(['cust_id','tran_date']).agg({'Return':'mean'})
```

```python
Cust_aggregate
```

# RFM scores

```python
RFM_scores = Cust_aggregate
```

```python
RFM_scores['R']    = pd.qcut(RFM_scores['Recency'], 3, labels=[3, 2, 1])
RFM_scores['F']    = pd.qcut(RFM_scores['Frequency'], 3, labels=[1, 2, 3])
RFM_scores['M']    = pd.qcut(RFM_scores['Monetary'], 3, labels=[1, 2, 3])
RFM_scores['A']    = pd.qcut(RFM_scores['Age'], 3, labels=[1, 2, 3])
RFM_scores['Re']    = pd.qcut(RFM_scores['Return'].rank(method='first'), 3, labels=[3, 2, 1])
```

```python
RFM_scores
```

```python
RFM_scores.info()
```

```python
RFM_scores['R'] = RFM_scores['R'].astype('int')
RFM_scores['F'] = RFM_scores['F'].astype('int')
RFM_scores['M'] = RFM_scores['M'].astype('int')
RFM_scores['A'] = RFM_scores['A'].astype('int')
RFM_scores['Re'] = RFM_scores['Re'].astype('int')
```

```python
RFM_scores.info()
```

## Sum all RFM scores

```python
RFM_scores['rfm_score'] = RFM_scores['R'] + RFM_scores['F'] + RFM_scores['M']+ RFM_scores['A']+ RFM_scores['Re']
```

```python
RFM_scores
```

```python
RFM_scores['rfm_score'].value_counts()
```

### RFM Describe

```python
RFM_scores.groupby('rfm_score')[['Recency','Frequency','Monetary','Age','Return']].describe().T
```

## Assign Label to RFM scores

1. 5 - 6 : At Risk
2. 7,8,9 : Need Attention
3. 10, 11, 12 : Potential Loyalist
4. 13,14,15 : Champion

```python
label = {5:'At Risk',
         6:'At Risk',
         7:'Need Attention',
         8:'Need Attention',
         9:'Need Attention',
         10:'Need Attention',
         11:'Potential Loyalist',
         12:'Potential Loyalist',
         13:'Champion',
         14:'Champion',
         15:'Champion'}
```

```python
RFM_scores['label'] = RFM_scores['rfm_score'].map(label)
```
